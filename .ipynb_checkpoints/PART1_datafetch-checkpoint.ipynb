````python
import requests
from bs4 import BeautifulSoup
import re
import time
import random

def get_page_source(url):
    try:
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/136.0.0.0 Safari/537.36',
            'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
            'Accept-Encoding': 'gzip, deflate, br',
            'Accept-Language': 'en-US,en;q=0.9',
            'Connection': 'keep-alive'
        }
        response = requests.get(url, headers=headers)
        response.raise_for_status()  
        response.encoding = "utf-8"
        return response.text
    except requests.RequestException as e:
        print(f"Error fetching {url}: {e}")
        return None
````

## 生成球队字典

````python
url1 = 'https://www.transfermarkt.com/premier-league/startseite/wettbewerb/GB1'
url2 = 'https://www.transfermarkt.com/laliga/startseite/wettbewerb/ES1'
url3 = 'https://www.transfermarkt.com/bundesliga/startseite/wettbewerb/L1'
url4 = 'https://www.transfermarkt.com/serie-a/startseite/wettbewerb/IT1'
url5 = 'https://www.transfermarkt.com/ligue-1/startseite/wettbewerb/FR1'

top5_url = [url1, url2, url3, url4, url5]
top5_dict = {'Premier League':0, 'LaLiga':0, 'Bundesliga':0, 'Serie A':0, 'Ligue 1':0}

for url, key in zip(top5_url, top5_dict.keys()):
    html = get_page_source(url)
    print(key,'...')
    soup = BeautifulSoup(html, 'html.parser')
    table = soup.find("table", class_="items")
    rows = table.tbody.find_all('tr')
    team_dict = {}
    for r in rows:
        d = r.find("td", class_="hauptlink no-border-links")
        if d:
            a = d.find('a')
            if a:
                team_dict[d.text.strip()] = a['href']
    top5_dict[key] = team_dict
    time.sleep(random.uniform(1, 3))
    print('Done')
````

## 生成球员字典

````python
# 设计一个带失败重试机制的request

def get_page_source_r(url, max_retries=4, delay=20):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/136.0.0.0 Safari/537.36',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
        'Accept-Encoding': 'gzip, deflate, br',
        'Accept-Language': 'en-US,en;q=0.9',
        'Connection': 'keep-alive'
    }

    for attempt in range(1, max_retries + 1):
        try:
            response = requests.get(url, headers=headers, timeout=10)
            response.raise_for_status()
            response.encoding = "utf-8"
            return response.text

        except requests.exceptions.RequestException as e:
            print(f"Request failed (attempt {attempt}) for {url}: {e}")
            if attempt < max_retries:
                print(f"Waiting {delay}s before retrying...")
                time.sleep(delay)
            else:
                raise Exception(f"Failed to fetch {url} after {max_retries} retries.")
````

````python
player_dict = {'Premier League':0, 'LaLiga':0, 'Bundesliga':0, 'Serie A':0, 'Ligue 1':0}
site = 'https://www.transfermarkt.com'
````

````python
for league_name, league_dict in top5_dict.items():
#     if league_name == 'Premier League' or league_name == 'LaLiga':
#         continue
    print(league_name, ':')
    league_player_dict = {}  # 用字典储存一个联赛的所有球员
    for team, url in league_dict.items():
        print(f"\t{team}...")
#         print('\t', code, url)
        html = get_page_source_r(site+url)
        soup = BeautifulSoup(html, 'html.parser')
        table = soup.find('table', class_='items')
        rows = soup.select(".even, .odd")
        for r in rows:
            cell = r.find('td', class_='posrela')
            data = cell.find('td', class_='hauptlink')
            league_player_dict[data.text.strip()] = data.find('a')['href']
        time.sleep(random.uniform(8,10))
    player_dict[league_name] = league_player_dict

print("Done")
````

我们需要的这个菜单，它被封装在了 Web Components 的 Shadow DOM 中。这跟传统的 HTML DOM 不一样，常规的爬虫（如 requests、BeautifulSoup）无法访问 Shadow DOM 的内容，因为它是由浏览器运行 JavaScript 动态生成的，而且是“封装”起来的。

````python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options

options = Options()
options.binary_location = "C:\\Program Files\\Google\\Chrome\\Application\\chrome.exe"  # 实际路径
options.add_argument("--headless")
options.add_argument("window-size=1920x1080")
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36")

service=Service(executable_path='chromedriver.exe')

driver = webdriver.Chrome(service=service, options=options)
driver.get(site+url) 

# time.sleep(3)  # 等待页面加载
````

````python
print(driver.title)
print(driver.page_source)
````

````python
# 执行 JavaScript 访问 shadow DOM
# 假设 shadow host 是某个 svelte 元素
shadow_host = driver.find_element(By.CSS_SELECTOR, "tm-subnavigation")  # 先找 Shadow Host
shadow_html = driver.execute_script("return arguments[0].shadowRoot.innerHTML", shadow_host)
````

````python
print(BeautifulSoup(shadow_html).prettify())
````

内容确实在 Shadow DOM 中，但只看到了“骨架屏”
也就是 `<div class="skeleton-item">` 是“加载中占位内容”，真正的内容是通过 JavaScript 动态渲染进去的。
